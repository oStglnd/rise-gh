# Histogram/Distribution plots for airflows

## Import dependencies and data

```python
# import dependencies
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
```

```python
# define relevant paths
home_path = os.path.dirname(os.getcwd())
data_path = home_path + '\\data\\'
plot_path = home_path + '\\plotting\\plots\\'
```

```python
# set seaborn style/theme
sns.set_theme()
sns.set_style('white')
```

```python
# get merged data
data = pd.read_csv(
    data_path + 'data_merged.csv',
    header=[0, 1],
    index_col=[0, 1, 2, 3]
)

# convert index.date col to datetime
#data.index = pd.to_datetime(data.index.values)
data.index = data.index.set_levels(
    levels=pd.to_datetime(data.index.get_level_values(3).values),
    level=3
)
```

```python
# keep only flow
data = data[[
    'flow'
]].copy()

# filter out variables
data = data.filter(
    regex='DC_SP.*_absolute|TA01_GP101|TA01_SP102.*_absolute', 
    axis=1
)
```

```python
# change names and rearrange columns
data = data.rename(
    columns = lambda col: col.replace('DC_', '').replace('_absolute', '').replace('_default', '')
)

```

## Plot distributions for airflows

```python
data.mean()
```

```python
#for month in [12, 1, 2]:
#    mask = data.index.get_level_values(0) == month
#    subset = data[mask]
# subset = data.flow[[
#     'Outdoor (SMHI)',
#     'Data center',
#     'Greenhouse'
# ]]

subset = data.flow[[
    'TA01_GP101',
    #'TA01_SP102',
    #'SP105',
    #'SP106',
    #'SP107',
    #'SP110',
    #'SP111',
    #'SP112'
]].copy()

# drop all index levels except date
subset = subset.droplevel(
    level=[0, 1, 2], 
    axis=0
)

ax = sns.histplot(
    data=subset, 
    stat='count', 
    binwidth=100, 
    binrange=[1000,5000],
    kde=False,
    legend='full',
    palette='muted',
    alpha=1.0
)

sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel('Airflow, $l/s$')
plt.ylabel('Count', loc='center', rotation=0, fontsize=12, labelpad=30)
#plt.xlim(-30, 60)
#plt.ylim(0, 0.4)
plt.title('Airflow distributions')

fpath = plot_path + 'TA01_GP101_flowdist.png'
plt.savefig(fpath, bbox_inches='tight', dpi=500)
plt.show()
```

```python
data[('flow', 'SP_SUM')] = \
    data.flow.SP105.values \
    + data.flow.SP106.values \
    + data.flow.SP107.values \
    + data.flow.SP108.values \
    + data.flow.SP110.values \
    + data.flow.SP111.values \
    + data.flow.SP112.values


subset = data.flow[[
    'SP_SUM',
    'TA01_SP102',
    #'SP105',
    #'SP106',
    #'SP107',
    #'SP110',
    #'SP111',
    #'SP112'
]].copy()

subset = subset.rename(columns={
    'SP_SUM':'To OUT',
    'TA01_SP102':'To GH'
})

# drop all index levels except date
subset = subset.droplevel(
    level=[0, 1, 2], 
    axis=0
)

# ax = sns.histplot(
#     data=subset, 
#     stat='count', 
#     binwidth=1000, 
#     binrange=[1000,80000],
#     kde=False, 
#     legend='full',
#     palette='bright',
#     alpha=0.6
# )

ax = sns.kdeplot(
    data=subset,
    fill=True,
    #multiple='stack',
    legend='full',
    palette='Blues',
    alpha=0.8
)

sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel('Airflow, $l/s$')
plt.ylabel('Density', loc='center', rotation=0, fontsize=12, labelpad=30)
#plt.xlim(-30, 60)
#plt.ylim(0, 0.4)
plt.title('Distributions of Airflow from DC chamber')

fpath = plot_path + 'DC_chamber_flowdists.png'
plt.savefig(fpath, bbox_inches='tight', dpi=500)
plt.show()
```

```python
subset = data.flow[[
    #'SP_SUM',
    #'TA01_SP102',
    'SP105',
    'SP106',
    'SP107',
    'SP110',
    'SP111',
    'SP112'
]].copy()

# drop all index levels except date
subset = subset.droplevel(
    level=[0, 1, 2], 
    axis=0
)

ax = sns.kdeplot(
    data=subset,
    multiple='fill',
    legend='full',
    palette='Blues_r',
    alpha=1.0
)

sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel('Airflow, $l/s$')
plt.ylabel('Density', loc='center', rotation=0, fontsize=12, labelpad=30)
#plt.xlim(-30, 60)
#plt.ylim(0, 0.4)
plt.title('Distributions of Airflow from DC chamber')

fpath = plot_path + 'DC_SP_flowdists.png'
plt.savefig(fpath, bbox_inches='tight', dpi=500)
plt.show()
```
