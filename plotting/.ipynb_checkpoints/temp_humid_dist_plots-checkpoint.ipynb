# Histogram/Distribution plots for temp and humid (DC, GH & OUT)

## Import dependencies and data

```python
# import dependencies
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
```

```python
# define relevant paths
home_path = os.path.dirname(os.getcwd())
data_path = home_path + '\\data\\'
plot_path = home_path + '\\plotting\\plots\\'
```

```python
# set seaborn style/theme
sns.set_theme()
sns.set_style('white')
```

```python
# get merged data
data = pd.read_csv(
    data_path + 'data_merged.csv',
    header=[0, 1],
    index_col=[0, 1, 2, 3]
)

# convert index.date col to datetime
#data.index = pd.to_datetime(data.index.values)
data.index = data.index.set_levels(
    levels=pd.to_datetime(data.index.get_level_values(3).values),
    level=3
)
```

```python
# keep only temp & humidity
data = data[[
    'temperatures',
    'humidity',
]].copy()

# filter out variables
data = data.filter(
    regex='GM10X|DC_GT401|SMHI(?!_)', 
    axis=1
)
```

```python
# calculate absolute humidity

def abs_humid(temp, rel_humid):
    abs_humidity =  6.112 * np.exp(17.67 * temp / (temp + 243.5)) * rel_humid * 2.1674 / (273.15 + temp)
    return abs_humidity

cols = ['DC_GT401_GM401', 'TA01_GT10X_GM10X', 'SMHI']
for col in cols:
    data['abs_humidity', col] = abs_humid(
        data.temperatures[col], 
        data.humidity[col]
    )
    
```

```python
# change names and rearrange columns
data = data.rename(columns={
    'DC_GT401_GM401':'Data center',
    'TA01_GT10X_GM10X':'Greenhouse',
    'SMHI':'Outdoor (SMHI)'
})

```

## Plot distributions for temperatures
