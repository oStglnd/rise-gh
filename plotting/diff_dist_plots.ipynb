# Histogram/Distribution plots for 10-min difference in vars

## Import dependencies and data

```python
# import dependencies
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
```

```python
# define relevant paths
home_path = os.path.dirname(os.getcwd())
data_path = home_path + '\\data\\'
plot_path = home_path + '\\plotting\\plots\\'
```

```python
# set seaborn style/theme
sns.set_theme()
sns.set_style('white')
```

```python
# get merged data
data = pd.read_csv(
    data_path + 'data_merged.csv',
    header=[0, 1],
    index_col=[0, 1, 2, 3, 4]
)

# convert index.date col to datetime
#data.index = pd.to_datetime(data.index.values)
data.loc[:, ('time', 'date')] = pd.to_datetime(data.time.date)
```

```python
# keep only ...
data = data[[
    'humidity',
    'temperatures',
    'flow',
    'sun',
    'time'
]].copy()

# # filter out variables
# data = data.filter(
#     regex='DC_SP.*_absolute|TA01_GP101|TA01_SP102.*_absolute', 
#     axis=1
# )
```

```python
data = data[[
    ('humidity', 'TA01_GT10X_GM10X'),
    ('temperatures', 'TA01_GT401_GM401'),
    ('temperatures', 'DC_GT301_damped'),
    ('flow', 'TA01_GP101'),
    ('sun', 'gsi'),
    ('sun', 'gsi_deriv'),
    ('time', 'date')
]].copy()

data.columns = data.columns.droplevel(0)
data.index = data.date
del data['date']
```

## Check vars and normalize

```python
data.mean()
```

```python
data.max()
```

```python
data.min()
```

```python
for col in data.columns:
    data[col] = (data[col] - data[col].mean()) / data[col].std()
```

```python
data = data - data.shift(periods=20)
```

## Plot distributions for airflows

```python
subset = data[[
    'TA01_GT10X_GM10X',
    'TA01_GT401_GM401',
    'DC_GT301_damped',
]]

ax = sns.kdeplot(
    data=subset,
    fill=True,
    #multiple='stack',
    clip=[-1, 1],
    legend='full',
    palette='bright',
    alpha=0.5
)

sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel('Change')
plt.ylabel('Density', loc='center', rotation=0, fontsize=12, labelpad=30)
plt.xlim(-1, 1)
#plt.ylim(0, 0.4)
plt.title('Distributions f. 10-minute changes')

# fpath = plot_path + 'DC_chamber_flowdists.png'
# plt.savefig(fpath, bbox_inches='tight', dpi=500)
plt.show()
```
